```python
import pandas as pd
```

```python
coffee_raw = pd.read_json("../public/arabica_data_cleaned.json", orient="records")
coffee_raw
```

```python
coffee_raw["Variety"] = coffee_raw["Variety"].replace([''], "Other")
```

```python
coffee_raw = coffee_raw.rename(columns={"Country.of.Origin": "Country"})
```

```python
coffee_raw["Country"] = coffee_raw["Country"].replace(['Tanzania, United Republic Of'], "Tanzania").replace(['United States (Hawaii)'], "Hawaii")
```

```python
coffee_raw
```

```python
grouped = coffee_raw.groupby(["Country"]).count()
```

```python
top_countries = list(grouped["FIELD1"].nlargest(15).index)
top_countries
```

```python
coffee_filtered = coffee_raw[coffee_raw['Country'].isin(top_countries)]
```

```python
coffee_filtered
```

```python
coffee_filtered.to_json("../public/arabica_data_cleaned_top15.json", orient="records")
```
